**Selection Sort** is a simple comparison-based sorting algorithm.  
It works by **repeatedly finding the minimum element** from the unsorted portion of the list and placing it at the beginning.

### Steps:
1. Start with the first element of the list.
2. Search the entire list (or the unsorted portion) to find the smallest element.
3. Swap this smallest element with the element at the current position.
4. Move the boundary of the sorted portion one step to the right.
5. Repeat until the list is fully sorted.

### Example:
For the list `[29, 10, 14, 37, 13]`:
- Pass 1: Smallest element is `10`. Swap with `29` → `[10, 29, 14, 37, 13]`
- Pass 2: Smallest in the remaining portion is `13`. Swap with `29` → `[10, 13, 14, 37, 29]`
- Pass 3: Smallest is `14`. No swap needed → `[10, 13, 14, 37, 29]`
- Pass 4: Smallest is `29`. Swap with `37` → `[10, 13, 14, 29, 37]`
- Sorted list achieved.

````python
def selection_sort(data, key):
    arr = data[:]  # copy (shallow)
    n = len(arr)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            if arr[j][key] > arr[min_idx][key]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
````

````python
#your code here 
twitchdata=open("twitchdata-update.csv")
twitchdata=list(csv.DictReader(twitchdata))
for ch in [row for row in twitchdata]:
    ch["Watch time(Minutes)"] = int(ch["Watch time(Minutes)"])
watch_sorted = selection_sort(twitchdata, key="Watch time(Minutes)")[:5]
print("Top 5 Channels by Watch Time :")
for ch in watch_sorted:
    print(ch)
````

### Quesiton 4: Searching
- Implement a linear search function to find a channel by its name.

- Sort the dataset by ```Followers``` and implement a binary search to quickly find the number of followers for a given channel.

- Compare the number of steps taken by linear vs. binary search.

````python
def linear_search(data,name):
    number=0
    steps=0
    channels_names=[] 
    for ch in [row for row in data]:
        ch["Channel"] = str(ch["Channel"])
    for row in data:
        channels_names.append(row['Channel'])
    for channel in channels_names:
        steps+=1
        number+=1
        if channel == name :
            return f"The channel {name} is channel n°{number}. The search took {steps} steps."
    return f"The channel {name} was not found. The search took {steps} steps."
def binary_search(data,name):
    for ch in data:
        ch["Channel"] = str(ch["Channel"])
    sorted_data = __builtins__.sorted(data, key=lambda x: x['Followers'])
    steps = 0
    low, high = 0, len(sorted_data) - 1
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        mid_followers = sorted_data[mid]['Followers']
        if mid_followers == name:
            channel = sorted_data[mid]['Channel']
            return f"The channel {name} has {channel} followers. The search took {steps} steps."
        elif mid_followers < name:
            low = mid + 1
        else:
            high = mid - 1
    return f"The channel {name} was not found. The search took {steps} steps."
````

````python
#your code here 
twitchdata=open("twitchdata-update.csv")
twitchdata=list(csv.DictReader(twitchdata))

print(linear_search(twitchdata,'Locklear'))
print(binary_search(twitchdata,'Locklear'))
````
